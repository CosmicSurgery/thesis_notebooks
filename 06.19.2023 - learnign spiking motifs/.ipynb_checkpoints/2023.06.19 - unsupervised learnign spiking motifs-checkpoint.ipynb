```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.signal import savgol_filter
%matplotlib inline
```

```python
N_PG = 4
N_nrn = 20
l_PG = 71
l_trial = 1000
PG_fr = 5 # hz
nrn_fr = 30 # hz

np.random.seed(41)
```

```python
# for matplotlib
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
len(colors)
```

# For generating the labels and the frequency of the PG's in the generated data.

```python
time = np.arange(0,l_trial,1)
pg_events = []

for k in range(N_PG):
    pg = np.array([np.random.randint(1000) for t in time])
    pg = np.where(pg <PG_fr)[0]
    pg = pg[pg < 1000-l_PG]
    pg_events.append((pg,1+k*np.ones(pg.shape)))
    
```

```python
pg_events
```

```python
motifs = []
'''
Parameters to the fs gaussian
f(x)=ae^{-\frac{(x-b)^{2}}{2c^{2}}}
'''


for k in range(len(pg_events)):
    motif = []
    for n in range(N_nrn):
        fr = np.random.poisson(nrn_fr)
        nrn_events = np.array([np.random.randint(0,1000) for k in range(l_PG)])
        nrn_events = np.array(np.where(nrn_events < fr)[0])
        nrn_nm = n*np.ones(len(nrn_events))
        motif.append((nrn_events,nrn_nm))
    motifs.append(motif)
print(len(motifs),len(motifs[0]),len(motifs[0][1]))
motifs[0][1] # PG_num , N_nrn
```

```python
%matplotlib inline
fig, axes = plt.subplot_mosaic("ABCD;EEEE;FFFF", figsize=(12,10));


[axes['A'].scatter(motifs[0][n][0],motifs[0][n][1], marker = '|',s=100,color=colors[0]) for n in range(N_nrn)]
[axes['B'].scatter(motifs[1][n][0],motifs[1][n][1], marker = '|',s=100,color=colors[1]) for n in range(N_nrn)]
[axes['C'].scatter(motifs[2][n][0],motifs[2][n][1], marker = '|',s=100,color=colors[2]) for n in range(N_nrn)]
[axes['D'].scatter(motifs[3][n][0],motifs[3][n][1], marker = '|',s=100,color=colors[3]) for n in range(N_nrn)]

[axes['E'].scatter(pg[0],pg[1], marker = '|',s=3000) for pg in pg_events]

for i, pg in enumerate(pg_events):
    for pg_time in pg[0]:
        [axes['F'].scatter(pg_time+np.array(motifs[i][n][0]),motifs[i][n][1], marker = '|',s=100,color=colors[i], alpha=0.5) for n in range(N_nrn)]
```

```python
nrn_activity = []
for n in range(N_nrn):
        
    temp_times= np.array([])
    temp_labels = np.array([])
    
    for i, pg in enumerate(pg_events):
        for pg_time in pg[0]:
            
            temp_times = np.append(temp_times, pg_time + np.array(motifs[i][n][0]))
            temp_labels = np.append(temp_labels, np.array(motifs[i][n][1]))
    
    nrn_activity.append((temp_times, temp_labels))
    nrn_activity[-1][0].sort()
#     nrn_activity[-1][0] = np.unique(nrn_activity[-1][0])
print(len(nrn_activity),len(nrn_activity[0]))
```

# How many motifs can we detect supervised learning?

Let's make a model of a spiking neuron, response to some sort of a signal...

```python
nrn_param = {'V_th': -55.0, 'V_reset': -75.0, 'tau_m': 10.0, 'g_L': 10.0, 'V_init': -75.0, 'E_L': -75.0,  'dt': 0.1}
nrn_param
```

```python
# Parameters of the detector neuron

E_L = -75
V_init = -75
g_L = 10
tau_m = 10
V_reset = -80 # add a recovery period?
V_th = -55
gain=1

delays = np.random.randint(l_PG, size=N_nrn)
```

Psp could probably jus be represented as an instanteous spike in mem-potential...

```python
lif = lambda I, V : (-(V-E_L) + I/g_L )  * ( 1 / tau_m)
```

```python
v_mem = np.zeros(l_trial)
v_mem[0] = V_init
I = 100*(np.sin(0.05*np.arange(0,l_trial)) + np.sin(0.08*np.arange(0,l_trial))+1)
for t in range(l_trial):
    if t==0:
        pass
    else:
        if v_mem[t-1] >= V_th:
            v_mem[t] = V_reset
            v_mem[t-1] = 45
        else:
            v_mem[t] = v_mem[t-1] + lif(I[t],v_mem[t-1])
```

```python
fig, ax = plt.subplots()
ax.plot(v_mem)
ax2 =ax.twinx()
ax2.plot(I,color='orange',alpha=0.5)
plt.show()

```

make the lif code a bit more general...

```python
# nrn_activity_binary; 
def lif_response(input_events, T, nrn_param = None, weights = None, gain=1):
    # input_events = time * m  -> m is the number of input neurons
    # time = vector incrementing
    v_mem = np.ones(len(time))
    spikes = []
    V = V_init
    
    for t in range(T):
        I = np.sum(input_events[t,:])
        if t==0:
            v_mem[t] = V_init
        else:
            if v_mem[t-1] >= V_th:
                v_mem[t] = V_reset
                v_mem[t-1] = 45
                spikes.append(t-1)
            else:
                v_mem[t] = v_mem[t-1] + (-(V-E_L) + gain*I/g_L )  * ( 1 / tau_m)
        V=v_mem[t]
    return v_mem, spikes
```
