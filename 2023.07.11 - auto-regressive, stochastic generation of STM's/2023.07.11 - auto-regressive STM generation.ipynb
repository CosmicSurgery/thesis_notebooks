```python
import numpy as np
import matplotlib.pyplot as plt
```

online generation of spiking motifs...


```python
T = 1000
t = 0
p_I = 0.005
th = 1
v=0
    
```

```python
spikes = []
t=0
plt.figure()
v=0
while t<T:
    t +=1
    v+=np.random.normal(scale=p_I*10)+p_I
    if v>=th:
        v=0
        spikes.append(t)
    if t%10==0:
        plt.scatter(t,v,color='black')
plt.show()
print(len(spikes))
    
```

```python
def sm_I(t,spikes,drop_off):
    if len(spikes)==0:
        return 0
    else:
        np.sin(2*t-spikes[-1])*np.e**(-drop_off*t)
```

```python
spikes = []
t=0
plt.figure()
v=0
sm_I=0
drop_off=0
while t<T:
    t +=1
    v+=np.random.normal(scale=p_I*10)+p_I+sm_I(t,spikes,drop_off)
    drop_off -= np.e**-(drop_off)
    if v>=th:
        v=0
        spikes.append(t)
        drop_off+=1
        
        
    if t%10==0:
        plt.scatter(t,v,color='black')
plt.show()
print(len(spikes))
```
