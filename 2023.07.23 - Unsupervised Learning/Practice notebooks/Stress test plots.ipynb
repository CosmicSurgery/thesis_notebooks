```python
import os
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
filename = 'scan_stats.json'
```

```python
if os.path.isfile(filename):
    with open(filename, 'r') as results_file:
        test = json.load(results_file)
df = pd.DataFrame(test)
df
```

```python
df_filt = df.dropna(subset=['pct good SM detected'])
df_filt = df_filt.fillna(0)
df_filt
```

```python
total_spikes = df_filt['total_spikes'].tolist()
total_motif_appearances = [np.sum(k) for k in df_filt['motif_appearances'].tolist()]
total_spikes_per_motif = [np.sum(k) for k in df_filt['spikes_per_motif'].tolist()]
compute_times = df_filt['time'].tolist()
Ms = df_filt['M'].tolist()
Ns = df_filt['N'].tolist()
Ds = df_filt['D'].tolist()
Ts = df_filt['T'].tolist()
nrn_frs = df_filt['nrn_fr'].tolist()
background_noise_frs = df_filt['background_noise_fr'].tolist()
pct_good_SMs_detected = df_filt['pct good SM detected'].tolist()
all_SM_qualities = df_filt['all SM quality'].tolist()
SM_accs = df_filt['SM_acc'].tolist()
snrs = df_filt['snr'].tolist()
motif_appearances = df_filt['motif_appearances'].tolist()
spikes_per_motif = df_filt['spikes_per_motif'].tolist()
```

```python
snrs_flat, SM_acc_flat, motif_appearances_flat, spikes_per_motif_flat = [], [], [], []
for i in range(len(snrs)):
    snrs_flat.extend(snrs[i])
    SM_acc_flat.extend(SM_accs[i])
    motif_appearances_flat.extend(motif_appearances[i])
    spikes_per_motif_flat.extend(spikes_per_motif[i])
SM_acc_flat = np.delete(SM_acc_flat,np.where(np.array(snrs_flat) == np.inf))
motif_appearances_flat = np.delete(motif_appearances_flat,np.where(np.array(snrs_flat) == np.inf))
spikes_per_motif_flat = np.delete(spikes_per_motif_flat,np.where(np.array(snrs_flat) == np.inf))
snrs_flat = np.delete(snrs_flat,np.where(np.array(snrs_flat) == np.inf))    
```

```python
plt.figure()
regression = np.arange(0,np.max(snrs_flat))
plt.scatter(snrs_flat,SM_acc_flat,alpha=0.5,s=1)
plt.plot(regression*scipy.stats.linregress(snrs_flat,SM_acc_flat)[0] + scipy.stats.linregress(snrs_flat,SM_acc_flat)[1],color='red')
plt.show()
```
