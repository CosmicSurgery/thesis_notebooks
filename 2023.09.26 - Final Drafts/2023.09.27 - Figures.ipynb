```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
date = 'Sept27th-'
filename = ''.join((date,'scan_stats.json'))
filename_SPADE = ''.join((date,'scan_stats_SPADE.json'))
filename, filename_SPADE
```

```python
df = pd.read_json(filename)
df_SPADe = pd.read_json(filename_SPADE)
```

```python

df['mean_acc'] = [np.mean(k) for k in df['SM_acc'].tolist()]
```

```python
fix = []
for k in df_SPADE['SM_acc'].tolist():
    if k[0] is not None:
        fix.append(np.mean(k))
    else:
        fix.append(np.nan)
df_SPADE['mean_acc'] = fix
```

```python
fig,ax = plt.subplots(1,3,figsize=(12,3))
plt.subplots_adjust(wspace=0.4)
ax[0].scatter(*df.iloc[:5][['M','mean_acc']].to_numpy().T, marker=',', alpha=0.5,label='HDP')
ax[0].scatter(*df_SPADE.iloc[:5][['M','mean_acc']].to_numpy().T, alpha=0.5,label='SPADE')
ax[0].set_xlabel('# of Patterns',fontsize=12)
ax[0].set_ylabel('Pattern accuracy',fontsize=12)
ax[0].set_xticks(df.iloc[:5]['M'])
ax[0].set_yticks(df.iloc[:5]['mean_acc'])
# ax[0].legend()

ax[1].scatter(*df.iloc[:5][['M','total_time']].to_numpy().T, marker=',', alpha=0.5,label='HDP')
ax[1].scatter(*df_SPADE.iloc[:5][['M','total_time']].to_numpy().T, alpha=0.5,label='SPADE')
ax[1].set_xlabel('# of Patterns',fontsize=12)
ax[1].set_ylabel('Compute time (s)',fontsize=12)
ax[1].set_xticks(df.iloc[:5]['M'])
ax[1].set_yticks(df_SPADE.iloc[1:4]['total_time'])
# ax[1].legend()

ax[2].scatter(*df.iloc[:5][['M','total_patterns']].to_numpy().T, marker=',', alpha=0.5,label='HDP')
ax[2].scatter(*df_SPADE.iloc[:5][['M','total_patterns']].to_numpy().T, alpha=0.5,label='SPADE')
ax[2].set_xlabel('# of Patterns',fontsize=12)
ax[2].set_ylabel('Patterns detected',fontsize=12)
ax[2].set_xticks(df.iloc[:5]['M'])
ax[2].set_yticks(df_SPADE.iloc[1:4]['total_patterns'])

ax[0].legend(bbox_to_anchor=(0.35, 0.3), loc='upper left')

plt.suptitle('M', fontsize=18)
```
