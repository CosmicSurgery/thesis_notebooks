```python
%matplotlib qt
alpha = 0.5
s = 10

fig,ax = plt.subplot_mosaic('A',figsize=(12,6))
ax['A'].scatter(df['total_spikes'],df['window_time'],alpha=alpha,s=s,label='Windowing')

x,y = df['total_spikes'],df['window_time']
sort_indexes  = sorted(range(len(x)), key=lambda k: x[k])
sorted_x = sorted(x)
sorted_y = [] 
for index in sort_indexes:
    sorted_y.append(y[index])

coeff = np.polyfit(sorted_x, sorted_y, 2)
line = np.linspace(sorted_x[0], sorted_x[-1], num = 100)
Polynomial = np.polyval(coeff, line)
ax['A'].plot(line,Polynomial)

ax['A'].scatter(df['total_spikes'],df['cluster_time'],alpha=alpha,s=s,label='Clustering')

x,y = df['total_spikes'],df['cluster_time']
sort_indexes  = sorted(range(len(x)), key=lambda k: x[k])
sorted_x = sorted(x)
sorted_y = [] 
for index in sort_indexes:
    sorted_y.append(y[index])

coeff = np.polyfit(sorted_x, sorted_y, 2)
line = np.linspace(sorted_x[0], sorted_x[-1], num = 100)
Polynomial = np.polyval(coeff, line)
ax['A'].plot(line,Polynomial)

ax['A'].scatter(df['total_spikes'],df['sequence_time'],alpha=alpha,s=s,label='Sequence Extracting')

x,y = df['total_spikes'],df['sequence_time']
sort_indexes  = sorted(range(len(x)), key=lambda k: x[k])
sorted_x = sorted(x)
sorted_y = [] 
for index in sort_indexes:
    sorted_y.append(y[index])

coeff = np.polyfit(sorted_x, sorted_y, 2)
line = np.linspace(sorted_x[0], sorted_x[-1], num = 100)
Polynomial = np.polyval(coeff, line)
ax['A'].plot(line,Polynomial)

ax['A'].scatter(df['total_spikes'],df['total_time'],alpha=alpha,s=s,label='Total')

x,y = df['total_spikes'],df['total_time']
sort_indexes  = sorted(range(len(x)), key=lambda k: x[k])
sorted_x = sorted(x)
sorted_y = [] 
for index in sort_indexes:
    sorted_y.append(y[index])

coeff = np.polyfit(sorted_x, sorted_y, 2)
line = np.linspace(sorted_x[0], sorted_x[-1], num = 100)
Polynomial = np.polyval(coeff, line)
ax['A'].plot(line,Polynomial)

plt.legend(fontsize=15)

plt.xlabel('Total Spikes',fontsize=24)
plt.ylabel('Time (s)', fontsize=24)
plt.title('Compute time for each stage of pattern detection',fontsize=30)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```

```python
from scipy.stats import pearsonr
import pandas as pd

def calculate_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 4)
    return pvalues
```

```python
df_pval = calculate_pvalues(df)
```
