```python
def fit(x,y, log = None,deg=2):
    idx = np.isfinite(x) & np.isfinite(y)
    x,y = x[idx],y[idx]

    sort_indexes  = sorted(range(len(x)), key=lambda k: x[k])
    sorted_x = sorted(x)
    sorted_y = [] 
    for index in sort_indexes:
        sorted_y.append(y[index])

    if log == 'semi-log':
        coeff = np.polyfit(sorted_x, np.log(sorted_y), deg)
    elif log == 'log-log':
        coeff = np.polyfit(np.log(sorted_x), np.log(sorted_y), deg)
    else:
        coeff = np.polyfit(sorted_x, sorted_y, deg)
    line = np.linspace(sorted_x[0], sorted_x[-1], num = len(x))
    Polynomial = np.polyval(coeff, line)
    return line, Polynomial
```

```python
from scipy.stats import pearsonr
import pandas as pd

def calculate_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 4)
    return pvalues
```
