```python
def fit(x,y, log = False,deg=2):
    idx = np.isfinite(x) & np.isfinite(y)
    x,y = x[idx],y[idx]

    sort_indexes  = sorted(range(len(x)), key=lambda k: x[k])
    sorted_x = sorted(x)
    sorted_y = [] 
    for index in sort_indexes:
        sorted_y.append(y[index])

    if log:
        coeff = np.polyfit(sorted_x, np.log(sorted_y), deg)
    else:
        coeff = np.polyfit(sorted_x, sorted_y, deg)
    line = np.linspace(sorted_x[0], sorted_x[-1], num = 100)
    Polynomial = np.polyval(coeff, line)
    return line, Polynomial
```
